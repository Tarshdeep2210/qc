```python
import cmath

z = 3 + 4j
print("Original Complex Number:", z)

polar_form = cmath.polar(z)

magnitude = polar_form[0]
phase = polar_form[1]

print("\nPolar Form:")
print("Magnitude (r):", magnitude)
print("Phase angle (theta in radians):", phase)

z_cartesian = cmath.rect(magnitude, phase)

print("\nConverted back to Cartesian form:")
print(z_cartesian)
```
